```python
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
# Tách DataFrame thành hai cột dựa trên điều kiện hàng chẵn và lẻ
ticker_df = df.iloc[::2].reset_index(drop=True)
name_df = df.iloc[1::2].reset_index(drop=True)
# Tạo DataFrame mới từ hai cột
merged_df = pd.DataFrame({
    "Ticket": ticker_df["DATA"],
    "Name": name_df["DATA"]
})

# In ra DataFrame mới đã ghép
merged_df.head(50)
```

```python
def process_name(name):
    for i, char in enumerate(name):
        if char.isdigit():
            if i > 3:  # Nếu ký tự số đầu tiên xuất hiện sau index 3
                return name[:i]  # Loại bỏ từ ký tự số đầu tiên này trở đi
            else:
                # Tìm ký tự số tiếp theo nếu nó tồn tại và xuất hiện trước index 3
                for j in range(i + 1, len(name)):
                    if name[j].isdigit() and j > 3:
                        return name[:j]
                break  # Dừng vòng lặp nếu không tìm thấy ký tự số thứ hai

    return name  # Trả về name ban đầu nếu không có ký tự số thứ hai

# Áp dụng hàm process_name cho cột "Name" của DataFrame merged_df
merged_df["Name"] = merged_df["Name"].apply(process_name)

print(merged_df.head(63))
```

```python
# Lưu DataFrame vào một tệp Excel
merged_df.to_excel("merged_data.xlsx", index=False)
```
